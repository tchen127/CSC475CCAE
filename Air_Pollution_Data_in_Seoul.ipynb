## Air pollution daily - sample plot for 1/1/2017 in Station 101


```python
pollutants = ['SO2', 'NO2', 'O3', 'CO', 'PM10', 'PM2.5']
```

```python
pNO2 = figure(title = 'sample graph for air pollutants for a 1/1/2017', x_axis_label = 'Hour')
pNO2.line(x = data1jan['Hour'], y = data1jan['NO2'], color = 'orange', legend_label = 'NO2')

show(pNO2)
```

```python
pCO = figure(title = 'sample graph for air pollutants for a 1/1/2017', x_axis_label = 'Hour',  y_axis_label = 'CO levels')
pCO.line(x = data1jan['Hour'], y = data1jan['CO'], color = 'green', legend_label = 'CO')

show(pCO)
```

```python
p = figure(title = 'sample graph for air pollutants for a 1/1/2017', x_axis_label = 'Hour')

p.line(x = data1jan['Hour'], y = data1jan['SO2'], color = 'red', legend_label = 'SO2')
p.line(x = data1jan['Hour'], y = data1jan['O3'], color = 'yellow', legend_label = 'O3')

show(p)
```

```python
pm = figure(title = 'sample graph for air pollutants for a 1/1/2017', x_axis_label = 'Hour', y_axis_label = 'SO2 levels')
pm.line(x = data1jan['Hour'], y = data1jan['PM10'], color = 'blue', legend_label = 'PM10')
pm.line(x = data1jan['Hour'], y = data1jan['PM2.5'], color = 'purple', legend_label = 'PM2.5')

show(p)
```
